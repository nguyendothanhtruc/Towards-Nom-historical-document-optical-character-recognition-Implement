```python
import shutil
import os
from glob import glob

FOLDER_NAME = "yolov5_0.4_conf_results"
lst_bbox = []
lst_folder = []
list_detected = glob("/content/yolov5/runs/detect/exp2/*", recursive = True)
for file in list_detected:
  if os.path.isfile(file):
    lst_bbox.append(file)
  else:
    lst_folder.append(file)

des_path = f"/content/gdrive/MyDrive/Intro_NLP/Code/Detection/{FOLDER_NAME}"
bbox_path = os.path.join(des_path, "bbox")
labels_path = os.path.join(des_path, "labels")
crop_path = os.path.join(des_path, "crops")

if not os.path.exists(bbox_path):
  os.makedirs(bbox_path)

for bbox in lst_bbox:
  file_name = bbox.split("/")[-1]
  shutil.copy(bbox, os.path.join(bbox_path, file_name))

shutil.copytree("/content/yolov5/runs/detect/exp2/labels", labels_path)

shutil.copytree("/content/yolov5/runs/detect/exp2/crops/word", crop_path)
```
